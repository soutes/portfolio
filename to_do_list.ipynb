```python
to_do_list = []
```

```python
def addTarefa():
    tarefa = input("Por favor digite a tarefa: ")
    to_do_list.append(tarefa)
    print(f"A tarefa '{tarefa}' foi adicionada na lista.")

def listaTarefa():
    if not to_do_list:
        print("Não existe item na Lista de tarefas.")
    else:
        print("Você possui os seguintes itens na sua lista:")
        for i, tarefa in enumerate(to_do_list):
            print(f"No índice = {i+1}. {tarefa}")

def delTarefa():
    listaTarefa()
    try:
        delTarefa = int(input("Digite o número da tarefa que deseja excluir."))-1
        if delTarefa >=0 and delTarefa < len(to_do_list):
            del to_do_list[delTarefa]
            print(f"A tarefa número {delTarefa+1} foi removida e o índice")
        else:
            print(f"A tarefa número {delTarefa} não foi encontrada.")
    except:
        print("Opção invalida")



if __name__ == "__main__":
   #Criando um loop para o programa rodar
   print("Bem-vindo ao meu script Tarafes a fazer :)")
   while True:
    print ("\n")
    print ("Por favor escolha uma das opções abaixo:")
    print ("___________________________________")
    print ("1. Adicionar uma nova tarefa.")
    print ("2. Excluir uma tarefa.")
    print ("3. Lista de tarefas")
    print ("4. Encerrar")

    selecao = input("Digite a opção desejada: ")
    if(selecao) == "1":
        addTarefa()
    elif(selecao) == "2":
        delTarefa()
    elif(selecao) == "3":
        listaTarefa()
    elif(selecao) == "4":
        break
    else:
        print("Opção invalida. Escolha uma das opções novamente.")

print("Até logo.")
```
